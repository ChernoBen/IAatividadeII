```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
```

```python
sb.pairplot(df)
```

```python
'''retirar as classes para que oa grupamento decida a qual classe elas pertencem'''
df
#recebendo as varidades/targets
X = np.array(df.drop('variety',axis= 1))
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
df['k-classes'] = kmeans.labels_
df
```

```python
sb.pairplot(df,hue='variety')
```

```python
sb.pairplot(df,'k-classes')
```
